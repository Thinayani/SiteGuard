## Download Dataset

Using the [Construction Site Safety](https://universe.roboflow.com/roboflow-universe-projects/construction-site-safety) dataset from Roboflow Universe.

> Get a free API key at [roboflow.com](https://roboflow.com) and paste it below.

```python
from roboflow import Roboflow

API_KEY = 'YOUR_ROBOFLOW_API_KEY'  # replace this

rf = Roboflow(api_key=API_KEY)
project = rf.workspace('roboflow-universe-projects').project('construction-site-safety')
version = project.version(28)
dataset = version.download('yolov8')
print('Downloaded to:', dataset.location)
```
